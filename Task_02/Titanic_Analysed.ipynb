#### Handling the missing values in each column 

```python
# Fill the age missing values with the mean of the age column
df['Age'] =df['Age'].fillna(df['Age'].mean())

# Fill the Fare missing values with the mean of the Fare column
df['Fare'] =df['Fare'].fillna(df['Fare'].mean())


# Fill the embarked column with the mode of the column
df['Embarked'] =df['Embarked'].fillna(df['Embarked'].mode()[0])


# Drop the Cabin column because there are too many missing values
df.drop(['Cabin'], axis=1, inplace=True)

df.isnull().sum()
```

#### Handling outliers

```python
# Box plot to show the outliers in the Fare paid by the passengers.
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fare'])
plt.show() 
```

```python

# Handle outliers using a box plot to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Fare'])
plt.title('Boxplot for Fare (Outlier Detection)')
plt.show()

# Outlier Handling using the interquartile range(IQR)
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Capping the outliers
df['Fare'] = np.where(df['Fare'] > upper_bound, upper_bound, df['Fare'])

print("Fare max value after capping:", df['Fare'].max()) 

```

#### Features engineering

```python
# Age Binning: Bucket the Age feature to categorize passengers by age into children, teenagers, adults, and seniors.
age_bins = [0, 12, 18, 55, 100]
age_labels = ['Child', 'Teenager', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
print(df[['Age', 'AgeGroup']].head())

# FamilySize: Adding SibSp and Parch to Form a FamilySize Feature.
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1 
print(df[['SibSp', 'Parch', 'FamilySize']].head())

# Fare Binning: Categorising the Fare feature into groups based on ticket price ranges.

# Using quartile-based binning for Fare:
fare_bins = df['Fare'].quantile([0, 0.25, 0.5, 0.75, 1]).tolist()
fare_labels = ['Low', 'Medium', 'High', 'Very_High']
df['FareGroup'] = pd.cut(df['Fare'], bins=fare_bins, labels=fare_labels, include_lowest=True)
print(df[['Fare', 'FareGroup']].head())
```

#### Machine  model using Logistic regression

```python
# Import machine learning libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Encode categorical variables in the dataset
label_cols = ['Sex', 'Embarked', 'AgeGroup', 'FareGroup']
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])
print(df[['Age', 'AgeGroup']].head(20))
```

```python
# Scale numerical features 
scaler = StandardScaler()
scale_cols = ['Age', 'Fare', 'FamilySize']
df[scale_cols] = scaler.fit_transform(df[scale_cols])
print(df[scale_cols].describe())
```

```python
# Features and target selection for the machine learning model
X = df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'FamilySize', 'AgeGroup', 'FareGroup']]
y = df['Survived']

# Split the data into training and testing sets, where 70% for training, 30% for testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Model training and testing
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Model evaluation 
print("Accuracy:", accuracy_score(y_test, y_pred))
print("n\Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
